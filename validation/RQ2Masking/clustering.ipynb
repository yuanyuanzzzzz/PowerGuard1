```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
import math

# =========================================================
# 1. Load and Merge Data
# =========================================================
df_feat = pd.read_csv("../../data/RQ2/background_traffic_full.csv")

# Load Performance Results
df_perf = pd.read_csv("../../data/RQ2/f1res.csv")
cols_to_clean = [c for c in df_perf.columns if "f1" in c.lower()]
for col in cols_to_clean:
    if df_perf[col].dtype == object:
        df_perf[col] = df_perf[col].str.replace("%", "").astype(float)

# Merge
df = pd.merge(df_feat, df_perf, on="device", how="inner")

# Calculate Impact (F1 Drop)
df["F1_Drop"] = df["f1_idle"] - df["f1_active"]

# =========================================================
# 2. Expanded Feature List (The "Suspects")
# =========================================================
suspects = {
    # --- Volume (The Control Group) ---
    "Volume (PPS)": "pps_active",
    "Background Traffic Strength (BTS)": "bts_active",
    
    # --- Complexity & Structure (The likely culprits) ---
    "Complexity (Entropy)": "flow_entropy_active",
    "Scattering (Num IPs)": "num_dst_ips_active",
    "Burstiness": "burst_count_active",
    "Size Uniformity (Kurt)": "size_kurt_active",
    "Rhythm Jitter (IAT Var)": "iat_var_active",
    
    # --- Specific Protocols ---
    "Discovery (SSDP)": "proto_ssdp_active", 
    "Discovery (mDNS)": "proto_mdns_active",
    "Encryption (TLS)": "proto_tls_active"
}

# =========================================================
# 3. Correlation Loop & Plotting
# =========================================================
# Setup grid: We have ~11 features, so a 3x4 grid works best
num_plots = len(suspects)
cols = 4
rows = math.ceil(num_plots / cols)

plt.figure(figsize=(24, 6 * rows))
sns.set_style("whitegrid")
plt.rcParams.update({'font.size': 14})

results = []

for i, (label, feature_col) in enumerate(suspects.items()):
    ax = plt.subplot(rows, cols, i+1)
    
    # check if feature exists in csv
    if feature_col not in df.columns:
        ax.text(0.5, 0.5, "Feature Not Found", ha='center', va='center')
        continue

    x_val = df[feature_col]
    y_val = df["F1_Drop"]
    
    # Clean NaNs
    mask = ~np.isnan(x_val) & ~np.isnan(y_val)
    x_clean = x_val[mask]
    y_clean = y_val[mask]
    
    if len(x_clean) < 2:
        continue
        
    r, p = pearsonr(x_clean, y_clean)
    
    # Scatter Plot with Regression
    # Color logic: Red if positive correlation (Masking driver), Blue if negative
    color = "#d62728" if r > 0 else "#1f78b4"
    
    sns.regplot(x=x_clean, y=y_clean, ax=ax, 
                scatter_kws={'s':80, 'alpha':0.6, 'edgecolor':'w'}, 
                line_kws={'color': color, 'linewidth': 2})
    
    # Styling
    ax.set_title(f"{label}\nR = {r:.2f} (p={p:.3f})", fontsize=16, fontweight='bold')
    ax.set_xlabel(feature_col, fontsize=12)
    ax.set_ylabel("F1 Drop (%)", fontsize=12)
    ax.grid(True, linestyle='--', alpha=0.5)
    
    results.append({
        "Feature Category": label, 
        "Correlation (r)": r, 
        "P-Value": p,
        "Significance": "*" if p < 0.05 else ""
    })

plt.tight_layout()
plt.savefig("rq2_expanded_correlation_analysis.pdf", dpi=300)
plt.show()

# =========================================================
# 4. Summary Table
# =========================================================
print("\n=== EXPANDED CORRELATION ANALYSIS ===")
res_df = pd.DataFrame(results).sort_values("Correlation (r)", ascending=False)
print(res_df.to_string(index=False))
```
